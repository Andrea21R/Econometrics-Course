Il CAPM intertemporale è dato da
$r_t = \alpha + \gamma E_{t-1}(\sigma^2_t)+u_t$ dove $r_t$ è il rendimento, $\sigma^{2}_t$ è la varianza condizionale e $U_t$ è il termine di disturbo. 
Il modello può essere stimato utilizzando i rendimenti di S&P 500 e una proxy
della varianza condizionale basata sul VIX $(h_t)$.

L'obiettivo è quello di stimare il parametro di avversione al rischio a partire dalla seguente regressione:
$r_t=\alpha+\gamma h_t+u_t$

```python
Y: pd.DataFrame = data.loc[:, ["returns"]]  
Y.head()
```

```python
X: pd.DataFrame = data.loc[:, ["impliedvix"]]  
X['constant'] = 1 
X: pd.DataFrame = X.loc[:, ["constant", "impliedvix"]]  
X.head()
```

```python
coeffs: pd.DataFrame = np.linalg.inv(X.transpose() @ X) @ X.transpose() @ Y
coeffs.index = X.columns  
coeffs
```

```python
Y_hat: pd.DataFrame = X @ coeffs
TSS: np.array = ((Y-np.mean(Y)).transpose() @ (Y-np.mean(Y))).values
ESS: np.array = ((Y_hat-np.mean(Y)).transpose() @ (Y_hat-np.mean(Y))).values
res: pd.DataFrame = Y - Y_hat
SSE: np.array = (res.transpose() @ res).values
R_2 = ESS/TSS
T: int = X.shape[0]
K: int = X.shape[1]
R_2_adj = 1 - (((T-1)/(T-K))*(1-R_2))
sig_2 = SSE/(T-K)
varcov: np.array = np.linalg.inv(X.transpose() @ X) * sig_2
standard_errors = np.sqrt(np.diag(varcov))
```

```python
coeffs["T-stat"] = coeffs["returns"].values / standard_errors
coeffs
```

Lo stimatore ha coefficienti BIAS e ciò è dovuto al fatto che $h_t$ è correlato con il termine di errore $u_t$. Una possibile soluzione è quella di stimare $\gamma$ con variabili strumentali e utilizzare come strumento $h_{t-1}$

```python
Y_2: pd.DataFrame = data.loc['03-Jan-90':'01-Jun-12',["returns"]] 
X_2: pd.DataFrame = data.loc['03-Jan-90':'01-Jun-12',["impliedvix"]]
X_2['constant'] = 1 
X_2: pd.DataFrame = X_2.loc[:, ["constant", "impliedvix"]] 
W: pd.DataFrame = data.loc['02-Jan-90':'31-May-12',["impliedvix"]] # Instrumental variable
W['constant'] = 1
W: pd.DataFrame = W.loc[:, ["constant", "impliedvix"]]
```

```python
N = Y_2.shape[0] #number of observations
k = X_2.shape[1]
df = N-k
print(df)
Y_2 = Y_2.to_numpy()
W = W.to_numpy()
X_2 = X_2.to_numpy()
```

```python
PWX = W @ np.linalg.inv(W.transpose() @ W) @ W.transpose() @ X_2
beta_2SLS = np.linalg.inv(PWX.transpose() @ PWX) @ PWX.transpose() @ Y_2
print(beta_2SLS)
```

```python
u_iv = Y_2 - X_2 @ beta_2SLS
s_hat = (u_iv.transpose() @ u_iv) / df 
var_hat = np.linalg.inv(PWX.transpose() @ PWX) * s_hat
stderr = np.sqrt(np.diag(var_hat))
t_stat_2sls = np.zeros(len(beta_2SLS))
t_stat_2sls[0] = beta_2SLS[0] / stderr[0]
t_stat_2sls[1] = beta_2SLS[1] / stderr[1]
```

Testare l'endogeneità di $h_t$. 

```python
X_3: pd.DataFrame = data.loc['03-Jan-90':'01-Jun-12',["impliedvix"]]
X_3 = X_3.to_numpy()
pi = np.linalg.inv(W.transpose() @ W) @ W.transpose() @ X_3
err2 = X_3 - W @ pi
X_4 = np.hstack((X_2, err2))
beta_2 = np.linalg.inv(X_4.transpose() @ X_4) @ X_4.transpose() @ Y_2
print(beta_2)
err3 = Y_2 - X_4 @ beta_2
sig2 = (err3.transpose() @ err3) / (Y_2.shape[0]-2)
V2 = np.linalg.inv(X_4.transpose() @ X_4)*sig2
V2
```
